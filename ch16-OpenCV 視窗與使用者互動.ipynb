### 16-2-6 程式範例：綜合進階視窗操作

```python
img = cv2.imread("sample/cat.jpg")
window_name = "Demo"

cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)  # 允許調整大小
cv2.imshow(window_name, img)
cv2.resizeWindow(window_name, 800, 600)          # 調整大小
cv2.moveWindow(window_name, 200, 100)            # 移動視窗
cv2.setWindowTitle(window_name, "New Title")     # 修改標題

cv2.setWindowProperty(window_name, cv2.WND_PROP_FULLSCREEN, 1) # 切換全螢幕
cv2.waitKey(3000)  # 等待 3 秒
cv2.setWindowProperty(window_name, cv2.WND_PROP_FULLSCREEN, 0) # 回復一般模式

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 16-3-4 程式範例：鍵盤輸入偵測與互動

```python
img_list = [
    cv2.imread("sample/cat.jpg"),
    cv2.imread("sample/dog.jpg"),
    cv2.imread("sample/bird.jpg")
]
index = 0
cv2.namedWindow("Album", cv2.WINDOW_AUTOSIZE)

while True:
    cv2.imshow("Album", img_list[index])
    key = cv2.waitKeyEx(0) # 使用 waitKeyEx() 捕捉特殊按鍵
    print(key)  # 顯示按鍵數值，方便除錯
    if key == ord('q') or key == 27: # 按 q 或 ESC 結束
        break
    elif key == ord('a') or key == 2424832: # 按 a 或左方向鍵，往左一張
        index = (index - 1) % len(img_list)
    elif key == ord('d') or key == 2555904: # 按 d 或右方向鍵，往右一張
        index = (index + 1) % len(img_list)
    elif key == ord('r'): # 按 r 重新回到第一張
        index = 0
cv2.destroyAllWindows()
```

### 16-4-3 程式範例：常見滑鼠事件類型

```python
img = cv2.imread("sample/cat.jpg")
# 滑鼠事件的 callback 函式
def mouse_event(event, x, y, flags, param):
    # event：事件型態
    # x, y：座標
    # flags：組合按鍵狀態（如 Ctrl/Shift 是否按下）
    # param：自訂傳遞參數
    if event == cv2.EVENT_LBUTTONDOWN:
        print(f"左鍵按下，座標=({x}, {y})")

# 綁定視窗與函式
window_name = "Mouse Demo"
cv2.namedWindow(window_name)
cv2.setMouseCallback(window_name, mouse_event)

while True:
    cv2.imshow(window_name, img)
    if cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()
```

### 16-4-4 程式範例：即時顯示滑鼠事件與座標

```python
def mouse_event(event, x, y, flags, param):
    if event == cv2.EVENT_MOUSEMOVE:
        print(f"滑鼠移動：({x}, {y})")
    elif event == cv2.EVENT_LBUTTONDOWN:
        print(f"左鍵按下：({x}, {y})")
    elif event == cv2.EVENT_LBUTTONUP:
        print(f"左鍵放開：({x}, {y})")
    elif event == cv2.EVENT_RBUTTONDOWN:
        print(f"右鍵按下：({x}, {y})")
    elif event == cv2.EVENT_RBUTTONUP:
        print(f"右鍵放開：({x}, {y})")
    elif event == cv2.EVENT_LBUTTONDBLCLK:
        print(f"左鍵雙擊：({x}, {y})")
    elif event == cv2.EVENT_RBUTTONDBLCLK:
        print(f"右鍵雙擊：({x}, {y})")

img = cv2.imread("sample/cat.jpg")
cv2.namedWindow("Mouse Demo")
cv2.setMouseCallback("Mouse Demo", mouse_event)

while True:
    cv2.imshow("Mouse Demo", img)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
cv2.waitKey(1) # 額外的 waitKey 可能有助於 MacOS 清理視窗
```

### 16-E-2 程式範例：顯示圖片並即時顯示座標

```python
IMAGE_PATH = 'sample/cat.jpg'
WINDOW_NAME = "Image Window"
mouse_coords = (0, 0)

# 定義滑鼠 Callback 函式（必須在 cv2.setMouseCallback 之前定義）
def show_coordinates(event, x, y, flags, param):
    global mouse_coords
    if event == cv2.EVENT_MOUSEMOVE:
        mouse_coords = (x, y)

# 讀取與初始化資料
original_img = cv2.imread(IMAGE_PATH)
height, width, _ = original_img.shape
resolution_text = f"Resolution: {width} x {height}"

# 建立視窗並綁定
cv2.namedWindow(WINDOW_NAME)
cv2.setMouseCallback(WINDOW_NAME, show_coordinates)

# 即時更新迴圈
while True:
    display_img = original_img.copy() # 產生副本，避免文字重疊
    
    # 繪製滑鼠座標
    coords_text = f"Position: ({mouse_coords[0]}, {mouse_coords[1]})"
    cv2.putText(display_img, coords_text, (15, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # 繪製圖片解析度（放在左下角）
    cv2.putText(display_img, resolution_text, (15, height - 20), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)
    
    cv2.imshow(WINDOW_NAME, display_img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
