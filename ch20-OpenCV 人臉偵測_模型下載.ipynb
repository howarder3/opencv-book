### 20-2-2 準備 Haar Cascade 人臉偵測模型

1. 從 opencv 的安裝路徑中尋找

```python
import cv2
import os

cascade_path = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"

if os.path.exists(cascade_path):
    print(f"分類器路徑：\n{cascade_path}")
else:
    print("找不到檔案！")
```

2. 透過網路下載 haarcascade_frontalface_default.xml

```python
import os
import requests

# 建立資料夾與設定路徑
os.makedirs("models", exist_ok=True)
cascade_path = "models/haarcascade_frontalface_default.xml"
url = "https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml"

# 檢查檔案是否存在，若無則進行下載
if not os.path.exists(cascade_path):
    print("正在下載 Haar Cascade 模型檔案...")
    response = requests.get(url)
    
    # 寫入檔案
    with open(cascade_path, 'wb') as f:
        f.write(response.content)
    print("下載完成！")
else:
    print("模型檔案已存在，不需重複下載。")
```

### 20-3-1 SSD 模型準備

* 透過網路下載 caffe SSD 人臉偵測模型

```python
# 安裝 tqdm, 可透過終端機或直接透過以下指令在此處快速安裝
!pip install tqdm
```

```python
import os
import requests
from tqdm import tqdm

# 建立資料夾
os.makedirs("models", exist_ok=True)

# 模型檔案清單與對應的下載網址
models = {
    "res10_300x300_ssd_iter_140000.caffemodel": "https://github.com/opencv/opencv_3rdparty/raw/dnn_samples_face_detector_20170830/res10_300x300_ssd_iter_140000.caffemodel",
    "deploy.prototxt": "https://raw.githubusercontent.com/opencv/opencv/master/samples/dnn/face_detector/deploy.prototxt"
}

# 迴圈檢查並下載檔案
for filename, url in models.items():
    filepath = os.path.join("models", filename)
    
    if os.path.exists(filepath):
        print(f"檔案已存在，跳過下載：{filename}")
        continue

    print(f"正在下載：{filename}")
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get('content-length', 0))
    
    # 結合 tqdm 顯示下載進度條，並自動管理資源關閉
    with open(filepath, 'wb') as f, tqdm(total=total_size, unit='iB', unit_scale=True) as bar:
        for data in response.iter_content(1024):
            f.write(data)
            bar.update(len(data))
```
