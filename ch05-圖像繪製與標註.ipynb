```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from opencv_tools import opencv_tools # 匯入封裝的功能
```

### 5-1-2 程式範例：在圖上畫點

```python
# 建立一張黑色圖片
img = np.zeros((400, 400, 3), dtype=np.uint8)

# 在圖片上畫點，注意 img 本身會被修改
cv2.circle(img, (200, 200), 2, (0, 0, 255), thickness=-1)
opencv_tools.show_img_by_matplotlib(img)
```

### 5-2-1 程式範例：在圖上畫圓

```python
# 建立一張黑色圖片
img = np.zeros((400, 400, 3), dtype=np.uint8)

# 畫一個紅色的實心圓
cv2.circle(img, (100, 100), 50, (0, 0, 255), thickness=-1)
# 畫一個綠色的空心圓 
cv2.circle(img, (150, 150), 100, (0, 255, 0), thickness=3)
# 畫一個藍色的空心圓 (使用抗鋸齒線)
cv2.circle(img, (200, 200), 100, (255, 0, 0), thickness=3, lineType=cv2.LINE_AA)

# 1-E 顯示圖片
opencv_tools.show_img_by_matplotlib(img)
```

### 5-3-2 程式範例：在圖上畫線

```python
# 建立一張黑色圖片
img = np.zeros((400, 400, 3), dtype=np.uint8)

# 畫一條綠色直線（左上到右下）
cv2.line(img, (0, 0), (399, 399), (0, 255, 0), thickness=10, lineType=cv2.LINE_AA)
# 畫一條藍色直線（右上到左下）
cv2.line(img, (399, 0), (0, 399), (255, 0, 0), thickness=5)

opencv_tools.show_img_by_matplotlib(img)
```

### 5-4-2 程式範例：在圖上畫矩形

```python
# 建立一張黑色圖片
img = np.zeros((400, 400, 3), dtype=np.uint8)

# 畫一個綠色的空心方形
cv2.rectangle(img, (50, 50), (350, 150), (0, 255, 0), thickness=3, lineType=cv2.LINE_AA)

# 畫一個紅色的實心方形
cv2.rectangle(img, (100, 200), (300, 350), (0, 0, 255), thickness=-1)

opencv_tools.show_img_by_matplotlib(img)
```

### 5-5-2 程式範例：在圖上畫橢圓

```python
# 建立一張黑色圖片
img = np.zeros((400, 400, 3), dtype=np.uint8)

# 畫綠色空心旋轉橢圓 (旋轉30度, 厚度3, 抗鋸齒)
cv2.ellipse(img, (200, 200), (100, 50), 30, 0, 360, (0, 255, 0), thickness=3, lineType=cv2.LINE_AA)
# 畫藍色實心橢圓
cv2.ellipse(img, (100, 300), (60, 30), 0, 0, 360, (255, 0, 0), thickness=-1)

opencv_tools.show_img_by_matplotlib(img)
```

### 5-5-3 程式範例：在圖上畫圓弧

```python
# 建立一張黑色圖片
img = np.zeros((400, 400, 3), dtype=np.uint8)

# 畫紅色半圓弧（180度到360度，對應圓的上半部）
cv2.ellipse(img, (200, 200), (100, 100), 0, 180, 360, (0, 0, 255), thickness=3, lineType=cv2.LINE_AA)
# 畫黃色橢圓弧（傾斜45度，取0度到180度的弧形範圍）
cv2.ellipse(img, (280, 280), (80, 40), 45, 0, 180, (0, 255, 255), thickness=3, lineType=cv2.LINE_AA)

opencv_tools.show_img_by_matplotlib(img)
```

### 5-6-3 程式範例：在圖上畫多邊形

```python
# 建立一張黑色圖片
img = np.zeros((400, 400, 3), dtype=np.uint8)

# 定義多邊形頂點座標（必須為 int32 型態）
pts1 = np.array([(50, 50), (300, 80), (350, 300), (100, 350)], np.int32)
# 轉換形狀為 (頂點數, 1, 2)
pts1 = pts1.reshape((-1, 1, 2))

# 畫黃色多邊形輪廓
cv2.polylines(img, [pts1], isClosed=True, color=(0, 255, 255), thickness=3, lineType=cv2.LINE_AA)

# 定義另一組多邊形頂點（必須為 int32 型態）
pts2 = np.array([(200, 150), (250, 200), (220, 300), (180, 250)], np.int32)
pts2 = pts2.reshape((-1, 1, 2))

# 畫填滿的青色多邊形
cv2.fillPoly(img, [pts2], color=(255, 255, 0))

opencv_tools.show_img_by_matplotlib(img)
```

### 5-7-2 程式範例：在圖上寫字

```python
# 建立一張白色圖片
img = np.ones((500, 1000, 3), dtype=np.uint8) * 255

# 字型選項與名稱配對
fonts = [
    (cv2.FONT_HERSHEY_SIMPLEX, 'FONT_HERSHEY_SIMPLEX'),
    (cv2.FONT_HERSHEY_PLAIN, 'FONT_HERSHEY_PLAIN'),
    (cv2.FONT_HERSHEY_DUPLEX, 'FONT_HERSHEY_DUPLEX'),
    (cv2.FONT_HERSHEY_COMPLEX, 'FONT_HERSHEY_COMPLEX'),
    (cv2.FONT_HERSHEY_TRIPLEX, 'FONT_HERSHEY_TRIPLEX'),
    (cv2.FONT_HERSHEY_COMPLEX_SMALL, 'FONT_HERSHEY_COMPLEX_SMALL'),
    (cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 'FONT_HERSHEY_SCRIPT_SIMPLEX'),
    (cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 'FONT_HERSHEY_SCRIPT_COMPLEX')
]

# 逐行寫上各種字型
y = 50
for font, name in fonts:
    cv2.putText(img, name, (30, y), font, 1.5, (0, 0, 0), 2, cv2.LINE_AA)
    y += 55 # 往下移動 55 像素

opencv_tools.show_img_by_matplotlib(img)
```

### 5-8-2 程式範例：在圖上標記特殊符號

```python
# 建立一張黑色背景的影像
img = np.zeros((300, 350, 3), dtype=np.uint8)

# 在不同位置畫上各種標記
cv2.drawMarker(img, (50, 50), (0, 0, 255), markerType=cv2.MARKER_CROSS, markerSize=30, thickness=2)
cv2.drawMarker(img, (100, 50), (0, 255, 0), markerType=cv2.MARKER_TILTED_CROSS, markerSize=30, thickness=2)
cv2.drawMarker(img, (150, 50), (255, 0, 0), markerType=cv2.MARKER_STAR, markerSize=30, thickness=2)
cv2.drawMarker(img, (200, 50), (0, 255, 255), markerType=cv2.MARKER_SQUARE, markerSize=30, thickness=2)
cv2.drawMarker(img, (250, 50), (255, 255, 0), markerType=cv2.MARKER_DIAMOND, markerSize=30, thickness=2)

opencv_tools.show_img_by_matplotlib(img)
```
