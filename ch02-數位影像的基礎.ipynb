### 2-3-3 彩色圖片的顏色通道

```python
# 建立黑色背景
width, height = 400, 400
background = np.zeros((height, width, 3), dtype=np.uint8)

# 定義圓的參數
radius = 100
center_red = (150, 150)  # 紅色圓心
center_green = (250, 150)  # 綠色圓心
center_blue = (200, 250)  # 藍色圓心

# 創建單獨的紅、綠、藍通道
red_layer = np.zeros_like(background)
green_layer = np.zeros_like(background)
blue_layer = np.zeros_like(background)

# 畫出紅色、綠色和藍色的圓
cv2.circle(red_layer, center_red, radius, (0, 0, 255), -1)  # 紅色 (B=0, G=0, R=255)
cv2.circle(green_layer, center_green, radius, (0, 255, 0), -1)  # 綠色 (B=0, G=255, R=0)
cv2.circle(blue_layer, center_blue, radius, (255, 0, 0), -1)  # 藍色 (B=255, G=0, R=0)

# 混合三層，調整透明度讓白色更明顯
blended_image = cv2.addWeighted(red_layer, 1.0, green_layer, 1.0, 0)
blended_image = cv2.addWeighted(blended_image, 1.0, blue_layer, 1.0, 0)

# 顯示結果
opencv_tools.show_img_by_matplotlib(blended_image) # 使用封裝好的 matplotlib 顯示圖片功能
```
